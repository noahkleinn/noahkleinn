```python
import sklearn as sk
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib as mlt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
%matplotlib inline
```

```python
df = pd.read_csv('PlayerPlayingPositionData.csv') # I selected the dataset because I think it would be funny to investigate into a game that I used to play
df.head()
```

```python
df.count() #show frequency of values
```

```python
df = df[['Preferred Positions', 'CB', 'CM', 'LB', 'RB', 'CDM', 'CAM', 'ST']] #subsetting the positions I want to investigate further into.
df['Preferred Positions'].str.split(" ",expand=True,).iloc[:,0] #splitting Preferred Positions into two
df1 = df.dropna() #dropping NAN values for 'GK'
df1.head(10) #table containing values
```

```python
X = df1.loc[:,'CB':'ST']
y = df1['Preferred Positions']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn import datasets
rf = RandomForestClassifier(random_state=1, n_estimators=100) #RF is a random algorithm, so to get the same results we need to use random_state
rf = rf.fit(X_train, y_train)
```

```python
rf.score(X_test,y_test) #the test predicted 33.93% correct. Given the fact that there are 7 variables in play, a succesrate of 33% is rather high.
```

Now it's time to evaluate the model by looking at the accuracy, precision and recall. I will be using a confusion matrix to calculate these values.


```python
rf.score(X, y, sample_weight=None) #Return the mean accuracy on the given test data and labels.
```

```python
# Make predictions for the test set
y_pred = rf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred)) #y_test, so not the real y yet
```

The precision appears to be 28%, which means it's higher then 1/4th chance the prediction will be right. With 34% however, recall is even higher than 1/3rd, which is rather high.

```python
# Get and reshape confusion matrix data
matrix = confusion_matrix(y_test, y_pred)

# Build the plot
plt.figure(figsize=(16,7))
sns.set(font_scale=1.4)
sns.heatmap(matrix, annot=True, annot_kws={'size':10},
            cmap=plt.cm.Greens, linewidths=0.2)

# Add labels to the plot
class_names = ['CB', 'ST', 'CAM', 
               'CDM', 'CM', 'RB',    
               'LB']
tick_marks = np.arange(len(class_names))
tick_marks2 = tick_marks + 0.5
plt.xticks(tick_marks, class_names, rotation=25)
plt.yticks(tick_marks2, class_names, rotation=0)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for Random Forest Model')
plt.show()
```

We can conclude that the test could be considered usable when predicting the preffered position for a new variable given the fact that the acuraccy, precision and recall are: 79,72%, 28% and 34%
